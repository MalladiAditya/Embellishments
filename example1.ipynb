# Image Denoising  in the Fourier domain

In this example, we will demonstrate how to denoise an image using the frequency domain 

### Initialization(importing necessary functions)

```python
import f
```

### Taking an input image
In this example, the image 'barbara.png' would be used.\
To use your own image, upload your image to the parent directory and replace the argument of the f.read() function used below with the name of your image

```python
img = f.read('barbara.png')
f.show_image(img)
```

### Creating the noise
The noise generated is a combination of low frequency sinusoids in the $x$ and $y$ directions.\
'$f_x$' is the frequency of the sinusoid in the $x$ direction.\
Similarly '$f_y$' is the frequency of the sinusoid in the $y$ direction.\
\
Note: Use small values for '$f_x$' and '$f_y$' as shown below

```python
fx = 5                                     # x-direction sinusoid frequency (Max value = Image width)
fy = 10                                    # y-direction sinusoid frequency (Max value = Image Height)
size = img.shape
noise  = f.sinusoid_noise(fx,fy,size)
```

Just checking the noise pattern generated

```python
f.show_image(noise)
```

Check: Count the number of wavelengths seperately in the $x$ and $y$ directions i.e how many times the black-white pattern is repeating.\
Answer: $f_x$ and $f_y$ times respectively   

### Adding Noise to the Image

```python
n_img  = img + 64*noise  # The value 32 is chosen arbitrarily
f.show_image(n_img)
```

After observing the above image, we can see that such noise is quite common. Thus, we need to find a way to process the above image to get an output resembling the initial input image(the uploaded image).

### Denoising
#### Oberving the Fourier Transform

The noise in the image can be clearly seen to be of sinusoidal nature. From the previous section, we have an idea of the noise frequency and thus, transforming the image into the frequency domain will be our first step.


```python
ft  = f.fourier(n_img)
```

The figure below is the magnitude spectrum

```python
%matplotlib notebook
f.show_image(f.mag(ft))
```

The above image apparently doesn't give a lot of details.\
So let us zoom a bit, around the center

```python
%matplotlib notebook
f.show_image(f.mag(ft),[240,280],[280,240])   # Zooming in to a small rectangle around the center 
#f.zoom(200,300,300,200)
```

Check: How would the magnitude spectrum of the image 'noise' look like? \
Answer: The 3 white dots as shown above, and the rest fully black

#### The Solution
Now that we recognize which frequency components are contributing to the noise, the simple solution is to just eliminate those components.\
In our example, those components would be the bright spots at (251,246) and (261,266) 

```python
%matplotlib inline
ft[246,251] = 0
ft[266,261] = 0
f.show_image(f.mag(ft),[240,280],[280,240])
```

Check: Is filtering involved in the above operation? If so, what filter has been applied?\
Answer: A notch filter (Very precise one though!!)

### Results 
Now to check the result of whatever we have done till now, take the inverse fourier transform of the above image 

```python
res = f.inv_fourier(ft)
f.show_image(res)
```

And we are done!!
