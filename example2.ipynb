# Is the Phase Spectrum of any use?
The Fourier Transform of an image in most cases turns out to be a complex quantity. This means that every image will have a non-trivial magnitude spectrum as well as a phase spectrum. In many filtering applications, the focus is mostly put on the magnitude spectrum. Everyone can plot magnitude responses of various filters, but struggle when asked about the phase response. This example demonstrates that the critical information of an image is actually contained in it's phase spectrum 

### Initializing setup

```python
import f
```

### Two input images
Images used here are of a lady and of a baboon.
Feel free to use your own images.\ 


Note: The images should be of the same size

```python
img1 = f.read('barbara.png')
img2 = f.read('baboonColor.png')
f.show_image(f.concat(img1,img2))
```

### Obtaining the Magnitude and Phase Spectra

```python
ft1 = f.fourier(img1)
ph1 = f.phase(ft1)           # Phase spectrum of 1st image
mag1 = abs(ft1)              # Magnitude spectrum of 1st image

ft2 = f.fourier(img2)
ph2 = f.phase(ft2)           # Phase spectrum of 2nd image    
mag2 = abs(ft2)              # Magnitude spectrum of 2nd image 
```

### Reconstruction
  
Firstly, let us reconstruct back our first image using both the magnitude as well as phase of the first image  

```python
res1 = f.inv_fourier(mag1*ph1)       # Mag 1 and Phase 1 
f.show_image(res1)
```

Now, we use the second image's phase spectrum('ph2') for our reconstruction  

```python
res2 = f.inv_fourier(mag1*ph2)      # Mag 1 and Phase 2 
f.show_image(res2)
```

There seems to a mistake, the recontructed output is nothing like our first image. Instead it closely resembles the second image!!\
Apparently, the phase spectrum holds much more information regarding important image characteristics than the magnitude spectrum

### Final Words
This simple examples illustrates the importance of the phase spectrum for any signal.\
